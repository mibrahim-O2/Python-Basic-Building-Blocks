Program that print 5 lines 1 2 3 4  5 in one line and sum all these values now the outer loop add the sum 4 times

```python
sum = 0
c2 = 1
while (c2 <= 4):
  c = 1
  while( c <= 5):
    sum = sum + c
    c = c + 1
c2 = c2 + 1
print("Final sum is : ",sum)
```
